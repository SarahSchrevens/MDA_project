# Retrieve data

```python
##import libraries request and bs4
import requests #library needed for requesting permission to website
from bs4 import BeautifulSoup #needed for filtering through website html code
```

```python
##adding url of site needed for data
url = "https://www.hetdepot.be/programma?page=1"
response = requests.get(url)
```

```python
##storing in beautifulsoup element for further usage
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find the element on the webpage
title = soup.select('h2[class = "calendar-item__title"]')
title
```

```python
for x in title:
    print(x.text)
```

```python
date = soup.select('h3') 
date
```

```python
for x in date:
    print(x.text)
```

# Create dataset (with date and title)

```python
import pandas as pd
```

```python
df = pd.DataFrame(title)
df.columns = ["event name"]
df
```

```python
df['date'] = date
df
```
