```python
##import libraries request and bs4
import requests #library needed for requesting permission to website
from bs4 import BeautifulSoup #needed for filtering through website html code

import pandas as pd
```

```python
##adding url of site needed for data
url = "https://www.weeronline.nl/Europa/Belgie/Leuven/4054147/weersverwachting-14dagen"
response = requests.get(url)
```

```python
##making sure everything works and the correct url is used
response.url
```

```python
##storing in beautifulsoup element for further usage
soup = BeautifulSoup(response.text, "html.parser")
```

```python
##select the class from the html code we are interested in, all events we need are of this type
dates = soup.select('div[class="styled__Day-sc-1g67f0q-15 cmEQpP"]') 

```

```python
##checking if correct data is scraped
for x in dates:
    print(x.text)
```

```python
##create df from scraped data
df_date = pd.DataFrame(dates)
df_date.columns = ["Date"]
df_date
```

```python
##scraping temperature data
temp = soup.select('div[class="styled__RowMinMaxTemp-sc-1g67f0q-8 dOyNoL"]') 
```

```python
##create list from scraped data about temperature
temperature = []
for y in temp:
    print(y.text)
    temperature.append(str(y.text))
```

```python
##create df from temperature data
df_temp = pd.DataFrame(temperature)
df_temp.columns = ["Temperature"]
df_temp
```

```python
##scrape rain data
rain = soup.select('span[class="styled__RowPrecipitation-sc-1g67f0q-9 dPVegl"]') 
```

```python
##create list from scraped rain data
rains=[]
for z in rain:
    print(z.text)
    rains.append(str(z.text))
```

```python
##create df from scraped rain data
df_rain = pd.DataFrame(rains)
df_rain.columns = ["Rain"]
df_rain
```

```python
##joining all df's
frames = [df_date, df_temp, df_rain]
df = pd.concat(frames, axis = 1, join = 'inner')
```
