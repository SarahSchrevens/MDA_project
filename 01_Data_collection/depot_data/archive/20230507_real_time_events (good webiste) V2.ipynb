# Retrieve data

```python
##import libraries request and bs4
import requests #library needed for requesting permission to website
from bs4 import BeautifulSoup #needed for filtering through website html code
import pandas as pd
```

```python
#page 1
```

```python
##adding url of site needed for data
url = "https://www.hetdepot.be/programma?page=0"
response = requests.get(url)
```

```python
##storing in beautifulsoup element for further usage
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find the element on the webpage
title = soup.select('h2[class = "calendar-item__title"]')
title
```

```python
for x in title:
    print(x.text)
```

```python
date = soup.select('h3')
date
```

```python
for x in date:
    print(x.text)
```

```python
df = pd.DataFrame(title[:30])
df.columns = ["event name"]
df
```

```python
df['date'] = date[:30]
df
```

```python
##adding url of site needed for data
url1 = "https://www.hetdepot.be/programma?page=1"
response1 = requests.get(url1)
```

```python
##storing in beautifulsoup element for further usage
soup1 = BeautifulSoup(response1.text, "html.parser")
```

```python
# Find the element on the webpage
title1 = soup1.select('h2[class = "calendar-item__title"]')
title1
```

```python
for x in title1:
    print(x.text)
```

```python
date1 = soup1.select('h3')
date1
```

```python
for x in date1:
    print(x.text)
```

```python
df1 = pd.DataFrame(title1[:30])
df1.columns = ["event name"]
df1
```

```python
# combine the content of the first and second rows in the "event name" column to match
df1['event name'] = df1['event name'].shift(-1)
df1 = df1.drop(df1.index[-1])
```

```python
# delete row 25
df1 = df1.drop(index=25)
#index reset
df1 = df1.reset_index(drop=True)
```

```python
df1['date'] = date1[:28]
df1
```
