# DATAFRAME WITH EVENT NAME AND DATE

```python
##import libraries request and bs4
import requests #library needed for requesting permission to website
from bs4 import BeautifulSoup #needed for filtering through website html code
import pandas as pd
```

```python
thislist1 = [] #empty list later used to store title
thislist2 = [] #empty list later used to store date

for i in range(0,11): #first 11 pages
    s = str(i) #converting i to a string, in order to concatenate it with the base URL string using the + operator in next step
    url = "https://www.hetdepot.be/archief?page=" + s #adding url of site needed for data 
    response = requests.get(url) #asking permission
    soup = BeautifulSoup(response.text, "html.parser") #store it into a soup object for further usage
    groups = soup.select('div.calendar__group') #scraping part containing date and title
    for group in groups:
        date = group.select_one('h3').text.strip() #get date
        items = group.select('div.calendar-item') #get title
        for item in items:
            title = item.select_one('h2.calendar-item__title').text.strip() #get actual title bc part of bigger block
            thislist1.append(title) #storing resulting titles in thislist1
            thislist2.append(date) #storing resulting dates in thislist2
#.text.strip() is used to extract the text content of the selected elements and remove leading or trailing whitespace
```

```python
#create a dataframe that have as column the info stored in the two lists resepctively
df_past = pd.DataFrame(thislist1)
df_past.columns = ["Event Name"] #column 1
df_past['Date'] = thislist2 #column 2
df_past
```

# Format 'date' column

Since I need to reformat the date comulmn and later split it in two columns startdate and enddate, I work with a data frame with only date, then create a for loop in which for each row i split the characters and extract only the necessary ones, and then I change the format into yyyy-mm-dd
