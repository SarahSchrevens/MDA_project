the task is simple. compare the noise data for several groups. the aim is to create a baseline, and the baselines will differ based on several possible categories. that being said, what categories actually make sense will depend on some experimentation from the group's point of view.

the first task: group the weather and noise aggregate data based on the 7 days of the week, and use visual comparison to check if the noise levels overlap with each other.

```python
import pandas as pd
```

```python
#load the data
dfs = []
# Loop through the filenames and read the dataframes
for i in range(1, 10):
    filename = f"df_{i:02d}.csv"  # Format the filename with leading zeros
    filepath = f"/Users/stefbp/Documents/GitHub/MDA_project/noise prediction model/Past Data matrix/{filename}"
    df = pd.read_csv(filepath)
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df[['date', 'day']] = df[['date', 'day']].apply(lambda x: pd.to_datetime(x)) #reading it from the raw file returns string dtypes for the time variables.
#variable time is left untouched.
```

```python
#dataframes based on the day of the week. have to create extra variable: night (20.00-06.00) or day (otherwise)
df['hour'] = df['date'].dt.hour
df['night'] = df['hour'].apply(lambda x: 1 if x >= 20 or x < 6 else 0)
```

```python
locs = df.location.unique()
locs
```

```python
import plotly.express as px
loc = locs[8]
naam = df[df['location']==loc]
fig = px.line(naam,x='date',y='laf50_per_hour',
              title='Naamsestraat ' + loc + ' Noise')
fig.update_traces(opacity=.4)
naam_night = naam[naam['night']==1];naam_day = naam[naam['night']==0]
fig.add_scatter(x=naam_night['date'],y=naam_night['laf50_per_hour'],
                mode="markers",marker=dict(size=5, color="red"))

fig.show()
```
