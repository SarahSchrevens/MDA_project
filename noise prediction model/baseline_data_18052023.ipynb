```python
import pandas as pd
#load the data
dfs = []
# Loop through the filenames and read the dataframes
for i in range(1, 10):
    filename = f"df_{i:02d}.csv"  # Format the filename with leading zeros
    filepath = f"/Users/stefbp/Documents/GitHub/MDA_project/noise prediction model/Past Data matrix/{filename}"
    df = pd.read_csv(filepath)
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df[['date', 'day']] = df[['date', 'day']].apply(lambda x: pd.to_datetime(x)) #reading it from the raw file returns string dtypes for the time variables.
#variable time is left untouched.
df['hour'] = df['date'].dt.hour
df['night'] = df['hour'].apply(lambda x: 1 if x >= 20 or x < 6 else 0)
```

```python
import numpy as np
df['weekend_status'] = pd.DataFrame(np.where(df['weekday'].isin(['Saturday', 'Sunday']), 1, 0))
```

36 baselines:
- 9 spots
- 2 categories of week (weekend/weekday)
- 2 categories of hours (night/day)

```python
baselines = df.groupby(['location', 'weekend_status', 'night'])['laf50_per_hour'].mean().reset_index()
baselines['weekend_status'] = baselines['weekend_status'].apply(lambda x: 'Weekend' if x == 1 else 'Weekday')
baselines['night'] = baselines['night'].apply(lambda x: 'Night' if x == 1 else 'Day')
```
